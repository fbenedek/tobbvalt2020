```python
#segítség: ezzel a kóddal URL-ekből tudtok képeket olvasni: így egyszerűbb nekünk is javítani

import numpy as np
import urllib
import cv2


def url_to_image(url):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    #Only made by Whirldata
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype='uint8')
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)

    # return the image
    return image
print (url_to_image('http://ichef.bbci.co.uk/onesport/cps/480/mcs/media/images/57210000/jpg/_57210683_57210682.jpg'))
```
